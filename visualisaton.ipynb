```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#getting the built in datasets os seaborn
sns.get_dataset_names()
```

```python
#loading the iris dataset
df=sns.load_dataset("iris")
```

```python
df.species.unique() #gives you unique value in categories
```

```python
df.species.value_counts() #counts on the basis of categories
```

```python
#line graph
import warnings
warnings.filterwarnings('ignore')
sns.lineplot(x='sepal_length',y='sepal_width',data=df,ci=None).set_title('line graph of sepal length v/s sepal width')

plt.show()
```

```python
#line plot
#hue--paramater can used on categorical data
sns.lineplot(x='sepal_length',y='sepal_width',data=df,hue='species',ci=None)
plt.show()
```

```python
# from the graph we can say that
#1) for setosa flower as sepal length increases ,sepal width also increses
#2) if sepal length is greater than 7 then it is virginica flower
#3) if sepal width is less than 2.45 ,then it is versicolor folwer
#4) setosa flower sepal width is more than any other species
```

```python
#continuous and categorical values
sns.lineplot(x='sepal_length',y='species',data=df,hue='species',ci=None)
plt.show()
```

```python
#form the graph we can say
#1) for setosa the sepal length ranges from 4.3 to 5.8
#2) for versocolor the sepal length varies from 4.9 to 7.2
#3) for virginica sepal length varies from 4.9 to 7.9
```

```python
sns.lineplot(x='petal_length',y='species',data=df,hue='species',ci=None)
plt.show()
```

```python
# Scatter plot (dot plot)
sns.scatterplot(x='sepal_length',y='sepal_width',data=df, hue='species')
plt.title("scatter plot")
plt.show()
```

```python
#insights
#1) for setosa we can see positive relation when sepal length is more sepal width also incraese
#2) for versicolor  and virginiaca there is no relationship between sepal length and sepal width
```

```python
#Bar plot (categorical estimate plot)
#A bar plot is basically used to aggregate the categorical data
# according to some methods and by default its the mean
# it can also be understand as a visualisation of the group by action
# to use this plot we choose categorical column
# the x axis and a numerical column for the y axis and e see that it
# craetes a plot taking a mean per categorical column

sns.barplot(x='species',y='sepal_width',data=df)
plt.show()
```

```python
# setosa fower sepal width is more than any other species
# versicolor is have lowest avg
```

```python
#count plot--value_count()--->for categoricla / discrete values
#a countplot basically counts the categories and returns a count of their occurences.
#df.species.value_counts()
sns.countplot(x='species',data=df)
plt.show()
```

```python
#Box plot
# A boxplot is sometimes known as box and whisker plot.it shows the distribution ofthe quatiative data
#that represents the comparisons between variable,boxplot shows the 
# shows the quartiles of the dataset while the whiskers extend to 
#show the rest of the distribution i.e., the dots indicating the 
#presence of outliers.
sns.boxplot(x='sepal_width',data=df)
plt.show()
```

```python
sns.boxplot(x='sepal_length',data=df)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_width',data=df)
plt.show()
```

```python
#violin plot
# it is similar to the boxplot except that is provides a 
#higher ,more advanced visualisation and uses the kernal
#density estimatiko  to give a better description about the
# data visualisation
```

```python
sns.violinplot(x='species',y='sepal_length',data=df)
plt.show()
```

```python
# insights
#for setosa most flower sepal length of 5
#for versicolor most flower sepal length of 5.5
# fro virginica most flower sepal length of 6.2
```

```python
#strip plot # it is basically creates a scatter plot based on the category
# it is created using stripplot()
sns.stripplot(x='species',y='petal_width',data=df ,hue='species')
plt.show()
```

```python
#histogram-frequency distribution of data
# A histogram is basically used to represent data provided
# in a form of some groups
# it is a accurate method for the graphical repesentation of numerical data
#distribution
sns.histplot(x='sepal_length',data=df)
plt.show()
```

```python
# histplot is basically on x-axis it takes weight and y-axis it takes count
```

```python
sns.histplot(x='sepal_width',data=df,hue='species')
plt.show()
```

```python
sns.histplot(x='sepal_width',data=df,kde=True)
plt.show()
#kde is kernal density estimation
#kde is apporximation of histogram of density curve
```

```python
# dist plot
#dist plot is used basically for univariant set of observations
#and visualizes it through a histogram
#i.e., only one observation and hence we choose one particular column of the dataset
sns.distplot(df['petal_width'],color='r',hist=False)# kde=false)
plt.show()
#DENSITY means the probability of given point to lie on the curve.
```

```python
#pair plot
#pair plot represnets pairwise relation across the entire dataframe and supports an

#addition argument called hue for categorical separation
#what it does basically is craete ajointplot between possible numerical
#column and takes a while if the dataframe is really huge

sns.pairplot(data=df,hue="species")
plt.show()
```

```python
#categorical plot - will show distribution of data by setting kind of property
# to a different plot
sns.catplot(x='petal_length',y='species',data=df,kind='swarm')  # swarm-strip plot
plt.show()
```

```python
# heat map
#heatmap is defined as a graphical represntation of data using colors to visualise the value of the 
#In this ,to represent more common values or higher activities brighter
#colors basically reddish colors are used and
#torepresent less common or activity values,
#darker colors are preferred

#correlation between the different parameters
data=df.drop('species',axis=1) #dropping the column since it is categorical column
tc=data.corr()  # to find the corrrelation only for continuous values
tc
```

```python
sns.heatmap(tc,annot=True)
plt.show()
```
